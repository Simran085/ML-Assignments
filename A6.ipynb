```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # Target classes (0, 1, 2)

# Train-Test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 1: Calculate the mean and variance for each feature per class
classes = np.unique(y_train)
means = {}
variances = {}
priors = {}

for cls in classes:
    X_class = X_train[y_train == cls]
    means[cls] = np.mean(X_class, axis=0)
    variances[cls] = np.var(X_class, axis=0)
    priors[cls] = X_class.shape[0] / X_train.shape[0]

# Step 2: Gaussian Probability Density Function
def gaussian_pdf(x, mean, var):
    exponent = np.exp(-((x - mean) ** 2) / (2 * var))
    return (1 / np.sqrt(2 * np.pi * var)) * exponent

# Step 3: Classify based on likelihood and prior
def predict(X):
    predictions = []
    for sample in X:
        posteriors = []
        for cls in classes:
            prior = np.log(priors[cls])  # log of prior
            likelihood = np.sum(np.log(gaussian_pdf(sample, means[cls], variances[cls])))  # log of likelihood
            posterior = prior + likelihood
            posteriors.append(posterior)
        predictions.append(np.argmax(posteriors))
    return np.array(predictions)

# Make predictions on the test set
y_pred_step_by_step = predict(X_test)

# Evaluate
print("Step-by-step Gaussian Naive Bayes Implementation:")
print("Accuracy:", accuracy_score(y_test, y_pred_step_by_step))
print("\nClassification Report:\n", classification_report(y_test, y_pred_step_by_step))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_step_by_step))
```
